```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/California_Houses.csv')
df.head()
```

```python
df.info()
df.shape
df.describe()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), 
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f',
            cbar_kws={'shrink': 0.8})
plt.title('California Houses - Correlation Matrix', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

df_new = df.drop(['Tot_Rooms', 'Tot_Bedrooms', 'Population', 'Households'], axis=1)
df_new['RoomsPerHousehold'] = df['Tot_Rooms'] / df['Households']
df_new['PopulationPerHousehold'] = df['Population'] / df['Households'] 
df_new['BedroomsPerRoom'] = df['Tot_Bedrooms'] / df['Tot_Rooms']
df_new['IncomePerRoom'] = df['Median_Income'] / df['Tot_Rooms']
df_new['MinDistanceToCity'] = df[['Distance_to_LA', 'Distance_to_SanDiego', 'Distance_to_SanJose', 'Distance_to_SanFrancisco']].min(axis=1)
df_new['LuxuryIndex'] = df['Tot_Rooms'] / df['Population']
df_new['Spaciousness'] = df['Tot_Rooms'] / df['Tot_Bedrooms']

new_features = ['IncomePerRoom', 'MinDistanceToCity', 'LuxuryIndex', 'Spaciousness']
correlations = df_new[new_features + ['Median_House_Value']].corr()['Median_House_Value'].sort_values(ascending=False)

print("New features correlations with price:")
print(correlations)

plt.figure(figsize=(14, 12))
sns.heatmap(df_new.corr(), 
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f')
plt.title('California Houses - Final Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()

top_features = df_new.corr()['Median_House_Value'].abs().sort_values(ascending=False).head(11)
print("\nTop 10 features by correlation with price:")
print(top_features)
```

```python
df_new['IncomePerHousehold'] = df['Median_Income'] / df['Households']
df_new['AvgDistanceToCities'] = df[['Distance_to_LA', 'Distance_to_SanDiego', 'Distance_to_SanJose', 'Distance_to_SanFrancisco']].mean(axis=1)
df_new['CoastalProximity'] = 1 / (1 + df['Distance_to_coast'])
df_new['AgeToIncomeRatio'] = df['Median_Age'] / df['Median_Income']
df_new['RoomEfficiency'] = df['Tot_Bedrooms'] / df['Tot_Rooms']
df_new['PopulationDensity'] = df['Population'] / ((df['Latitude'].max() - df['Latitude'].min()) * (df['Longitude'].max() - df['Longitude'].min()))

df_new['LatitudeSquared'] = df['Latitude'] ** 2
df_new['LongitudeSquared'] = df['Longitude'] ** 2
df_new['LatLonProduct'] = df['Latitude'] * df['Longitude']

df_new['LogIncome'] = np.log(df['Median_Income'])
df_new['LogDistanceToCoast'] = np.log(1 + df['Distance_to_coast'])

all_new_features = ['IncomePerHousehold', 'AvgDistanceToCities', 'CoastalProximity', 
                   'AgeToIncomeRatio', 'RoomEfficiency', 'PopulationDensity',
                   'LatitudeSquared', 'LongitudeSquared', 'LatLonProduct',
                   'LogIncome', 'LogDistanceToCoast']

correlations = df_new[all_new_features + ['Median_House_Value']].corr()['Median_House_Value'].sort_values(ascending=False)

print("All new features correlations with price:")
print(correlations)

best_features = correlations[abs(correlations) > 0.1].drop('Median_House_Value')
print(f"\nBest features (|correlation| > 0.1): {len(best_features)}")
print(best_features)

plt.figure(figsize=(20, 14))
sns.heatmap(df_new.corr(), 
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f')
plt.title('California Houses - Final Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
df_final = df.copy()

df_final['RoomsPerHousehold'] = df['Tot_Rooms'] / df['Households']
df_final['BedroomsPerRoom'] = df['Tot_Bedrooms'] / df['Tot_Rooms']

df_final['Spaciousness'] = df['Tot_Rooms'] / df['Tot_Bedrooms']
df_final['MinDistanceToCity'] = df[['Distance_to_LA', 'Distance_to_SanDiego', 'Distance_to_SanJose', 'Distance_to_SanFrancisco']].min(axis=1)
df_final['AgeToIncomeRatio'] = df['Median_Age'] / df['Median_Income']
df_final['AvgDistanceToCities'] = df[['Distance_to_LA', 'Distance_to_SanDiego', 'Distance_to_SanJose', 'Distance_to_SanFrancisco']].mean(axis=1)
df_final['RoomEfficiency'] = df['Tot_Bedrooms'] / df['Tot_Rooms']
df_final['LuxuryIndex'] = df['Tot_Rooms'] / df['Population']

df_final = df_final.drop(['Tot_Rooms', 'Tot_Bedrooms', 'Population', 'Households'], axis=1)

print(f"Final data shape: {df_final.shape}")
print(f"Columns: {df_final.columns.tolist()}")

plt.figure(figsize=(14, 12))
sns.heatmap(df_final.corr(), 
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f')
plt.title('California Houses - Final Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
df_final_clean = df_final.drop([
    'Distance_to_LA', 'Distance_to_SanDiego', 'Distance_to_SanJose', 'Distance_to_SanFrancisco',
    'Longitude', 'Median_Age'
], axis=1)

print(f"Cleaned shape: {df_final_clean.shape}")
print(f"Columns: {df_final_clean.columns.tolist()}")

plt.figure(figsize=(14, 12))
sns.heatmap(df_final_clean.corr(), 
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f')
plt.title('California Houses - Final Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```
