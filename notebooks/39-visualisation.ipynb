```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", font_scale=1.3)
```

```python

r2_scores = {
    "Decision Tree": 0.917,
    "Random Forest": 0.9098,
    "Gradient Boosting": 0.9412,
}

acc_scores = {
    "Decision Tree": 0.3217,
    "Random Forest": 0.3743,
    "Gradient Boosting": 0.3111,
}
```

```python

plt.figure(figsize=(8,5))
plt.title("Regression Models — R² Score Comparison")

models = list(r2_scores.keys())
values = list(r2_scores.values())

sns.barplot(x=models, y=values)
plt.ylim(0, 1)
plt.ylabel("R² Score")
plt.xlabel("Model")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.title("Classification Models — Accuracy Comparison")

models = list(acc_scores.keys())
values = list(acc_scores.values())

sns.barplot(x=models, y=values)
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.show()
```

```python

import pandas as pd

df = pd.DataFrame({
    "Model": list(r2_scores.keys()),
    "R²": list(r2_scores.values()),
    "Accuracy": list(acc_scores.values())
})

df
```
