```python
import torch
import pandas as pd
import os
root = "../data/fashion-mnist"

train_csv = os.path.join(root, "fashion-mnist_train.csv")
test_csv = os.path.join(root, "fashion-mnist_test.csv")

df_tr = pd.read_csv(train_csv)
df_te = pd.read_csv(test_csv)

df_tr.head()
```

```python
y_tr = torch.tensor(df_tr["label"].values, dtype=torch.long)
X_tr = torch.tensor(df_tr.drop(columns=["label"]).values, dtype=torch.float32)

y_te = torch.tensor(df_te["label"].values, dtype=torch.long)
X_te = torch.tensor(df_te.drop(columns=["label"]).values, dtype=torch.float32)


print(X_tr.shape, y_tr.shape)
print(X_te.shape, y_te.shape)

print(X_tr.min(), X_tr.max())
```

```python
X_tr = X_tr / 255.0
X_te = X_te / 255.0

print(X_tr.min(), X_tr.max())
```

```python
bundle = {
    "X_train": X_tr.float().cpu().contiguous(),
    "y_train": y_tr.long().cpu().contiguous(),
    "X_test":  X_te.float().cpu().contiguous(),
    "y_test":  y_te.long().cpu().contiguous(),
}
torch.save(bundle, os.path.join(root, "fashion_mnist_knn.pt"))
```
